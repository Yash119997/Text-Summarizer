```
from operator import itemgetter
from numpy import einsum, sum, dot, arccos, zeros, diag

from scipy.spatial import distance 

punctuations = ''',:';!'''

class Latent_Semantic_Analysis(object):

    def __init__(self, punctuations):
        self.punctuations = punctuations
        self.vcb = {}
        self.word_count = 0        

    def cal_word_count(self, cnn_article):
        cnn_article_words = cnn_article.split();
        for voc in cnn_article_words:

            voc = voc.translate(str.maketrans('','', self.punctuations))
            if voc in self.vcb:
                self.vcb[voc].append(self.word_count)
            else:
                self.vcb[voc] = [self.word_count]
        self.word_count += 1  

    def generate_matrix(self):

        self.freq_wds = []
        vcb_mat = self.vcb
        
        for w in vcb_mat.keys():
            l_vcb = len(vcb_mat[w])
            if  l_vcb > 1:
                self.freq_wds.append(w)
        lnt = len(self.freq_wds)
        self.freq_wds.sort()
        
        self.svd_mat = zeros([lnt, self.word_count])
        for x, y in enumerate(self.freq_wds):
            for w in vcb_mat[y]:
                self.svd_mat[x,w] += 1
        self.vcb = vcb_mat
    
    def Cal_SVD(self):
        self.MatrixU, self.MatrixS, self.SVDvT = matrix_svd(self.svd_mat)
        self.MatrixU = -1 * self.MatrixU
        self.SVDvT = -1 * self.SVDvT
```
